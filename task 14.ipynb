```python
import pandas as pd
```

```python
import numpy as np
```

```python


titanic = pd.read_csv('https://raw.githubusercontent.com/nagamounika5/Datasets/master/titanic.csv')
titanic

```

```python
titanic.isnull().sum()
```

```python
#Replace null values 'Age' column with ffill()
titanic1 = titanic.copy()
titanic1 = titanic[['Age']].ffill()
titanic1
```

```python
# After replacing the null values in the column of Age, there are no null values left
titanic1.isnull().sum()
```

```python
# Replacing the Cabin column with mode of that column
titanic['Cabin'].fillna(titanic['Cabin'].mode()[0], inplace=True)
titanic
```

```python
# Replacing the Embarked column with mode of that column
titanic['Embarked'].fillna(titanic['Embarked'].mode()[0], inplace=True)
titanic
```

```python
# At a time replacing the two column null values with the mode of their columns
columns = ["Cabin", "Embarked"]
titanic[columns] = titanic[columns].fillna(titanic.mode().iloc[0])
titanic
```

```python
# Only the Age column null values are left after replacing the Cabin and Embarked column null values
titanic.isnull().sum()
```
