```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
d_f = pd.read_csv("https://raw.githubusercontent.com/LavanyaPolamarasetty/Datasets/master/Classification/Social_Network_Ads.csv")
d_f
```

```python
plt.title("Line Plot")
plt.plot(d_f['Purchased'])
plt.show()
```

```python
plt.title("Scatter Plot")
plt.scatter(d_f["User ID"],d_f["Age"])
plt.ylabel("Age")
plt.xlabel("User ID")
plt.show()
```

```python
plt.title("Histogram")
plt.hist(d_f["Age"])
plt.xlabel("Age")
plt.show()
```

```python


plt.title("Bar Graph")
plt.bar(d_f["Gender"],d_f["EstimatedSalary"])
plt.xlabel("Gender")
plt.ylabel("EstimatedSalary")
plt.show()

```

```python
plt.title("Pie Chart")
plt.pie(d_f["EstimatedSalary"][:10],labels = d_f["User ID"][:10], autopct = "% 1.1f%%")
plt.show()

```

```python
plt.title("Box Plot")
plt.boxplot(d_f["EstimatedSalary"])
plt.show()
```

```python


fig, area = plt.subplots(2, 2)
area[0][0].plot(d_f['Purchased'])
area[0][1].plot(d_f['Gender'])
area[1][0].plot(d_f['Age'])
area[1][1].plot(d_f['EstimatedSalary'])
plt.show()

```

```python
plt.subplot(1, 2, 2)
plt.plot(d_f['Purchased'])
plt.subplot(1, 2, 1)
plt.plot(d_f['EstimatedSalary'])
plt.show()
```
